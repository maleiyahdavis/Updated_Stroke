Imports

```python
import pandas as pd
```

### Extract

Reading our Kaggle CSV file, let's look at our first 5 rows

```python
stroke_df = pd.read_csv("./healthcare-dataset-stroke-data.csv")
stroke_df.head(5)
```

### Data Transformation
We don't have extra or missing columns, let's take a look at our types and null counts

```python
stroke_df.info()
```

We have 5110 rows of data in our dataset looking through the columns we seem to only have null values in our bmi column

```python
stroke_df.loc[stroke_df["bmi"].isnull()].head(5)
```

We would like to use bmi for later analysis, so let's replace bmi with the average depending on gender

Let's find the calculate the mean with a groupby

```python
female_bmi = stroke_df.groupby("gender")["bmi"].mean().Female
male_bmi = stroke_df.groupby("gender")["bmi"].mean().Male
```

To find and replace null values depending on gender we'll make seperate dataframes for each gender and append together

```python
female_df = stroke_df.loc[stroke_df["gender"]=="Female"].fillna(female_bmi)
male_df = stroke_df.loc[stroke_df["gender"]=="Male"].fillna(male_bmi)
stroke_df = pd.concat([female_df, male_df])
stroke_df.head(10)
```

```python
print("Columns with null values: ")
stroke_df.isnull().sum()
```

### Looking better, no nulls

Our age column appears as a float type, but has no value in its first decimal, we can cast it back to being an int

```python
stroke_df = stroke_df.astype({"age":"int"})
stroke_df.dtypes
```

### Export to Analysis

```python
stroke_df.to_csv("./stroke.csv")
```
