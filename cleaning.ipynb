```python
import pandas as pd
```

```python
stroke_df = pd.read_csv("./healthcare-dataset-stroke-data.csv")
stroke_df.head(5)
```

```python
female_bmi = stroke_df.groupby("gender")["bmi"].mean().Female
male_bmi = stroke_df.groupby("gender")["bmi"].mean().Male
```

```python
stroke_df.loc[stroke_df["bmi"].isnull()]
```

```python
female_df = stroke_df.loc[stroke_df["gender"]=="Female"].fillna(female_bmi)
male_df = stroke_df.loc[stroke_df["gender"]=="Male"].fillna(male_bmi)
```

```python
stroke_df = pd.concat([female_df, male_df])
stroke_df
```

```python
stroke_df.dtypes
```

```python
stroke_df = stroke_df.astype({"age":"int"})
stroke_df.dtypes
```

```python
stroke_df.to_csv("./stroke.csv")
```
